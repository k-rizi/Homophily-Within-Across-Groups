```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# Probability distributions
triangle = [0.4977789916694153, 0.15071264998388104, 0.07344356723902098, 0.05760348168597331, 0.0727167766286929, 0.14774453279301666]
square = [0.030937582332237125, 0.06932680881826901, 0.1300870848580512, 0.2044020692117611, 0.26893946007004965, 0.296306994709632]
circle = [0.08218346400928819, 0.24814330813087346, 0.351743507674971, 0.2340749880969601, 0.07312894821129119, 0.010725783876616195]

# X-axis values
x = np.arange(6) + 0.3  # Shift points slightly to the right

# Set Seaborn style
sns.set(style="white")  # Clean white background, no grid

# Define an improved bluish-gray color palette
palette = ["#6E8192", "#4E5F6A", "#1A2528", "#36434A"]

# Create figure with adjusted aspect ratio
fig, ax = plt.subplots(figsize=(8, 6))


sns.lineplot(x=x, y=triangle, marker="^", markersize=14, markerfacecolor="white", markeredgewidth=2,
             markeredgecolor=palette[1], color=palette[1], linestyle="dashed", label=r"$h_5=0.6$, $n_r=0.3$", linewidth=2.5, ax=ax)

sns.lineplot(x=x, y=square, marker="s", markersize=14, markerfacecolor="white", markeredgewidth=2,
             markeredgecolor=palette[2], color=palette[2], linestyle="dashed", label=r"$h_5=0.2$, $n_r=0.7$", linewidth=2.5, ax=ax)

sns.lineplot(x=x, y=circle, marker="o", markersize=14, markerfacecolor="white", markeredgewidth=2,
             markeredgecolor=palette[3], color=palette[3], linestyle="dashed", label=r"$h_5=0.0$, $n_r=0.4$", linewidth=2.5, ax=ax)

# Improve labels with LaTeX formatting
#plt.xlabel(r"number of red nodes $(i)$", fontsize=20, labelpad=15)
#plt.ylabel(r"$F_6$", fontsize=20, labelpad=15)

# Adjust x-ticks so they align properly with the shifted points
plt.xticks(np.arange(6) + 0.3, labels=[f"{i}" for i in range(6)], fontsize=18)
plt.yticks(fontsize=18)

# Adjust axis limits for better spacing
plt.xlim(-0.2, max(x) + 0.5)  # Expanding left and right to balance the shift
plt.ylim(-0.02, max(triangle) + 0.05)

# Improve legend
plt.legend(fontsize=16, loc="upper center", frameon=False, ncol=1)

# Remove default spines
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_visible(False)

# Draw arrowed axes
ax.annotate("", xy=(max(x) + 0.5, 0), xytext=(-0.2, 0),
            arrowprops=dict(arrowstyle="->", linewidth=2, color="black"))
ax.annotate("", xy=(0, max(triangle) + 0.05), xytext=(0, -0.02),
            arrowprops=dict(arrowstyle="->", linewidth=2, color="black"))

# Save the figure with tight bounding box
plt.savefig("../results/1e.pdf", dpi=300, bbox_inches="tight")

# Show the plot
plt.show()
```
