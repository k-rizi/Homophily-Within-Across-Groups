```python
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize
from math import comb
import os, pickle
import itertools

def plot_(data_sets_names, label_map=None):
    plt.figure(figsize=(4, 7                    
                       ))  # Increase figure size for better visibility and separation

    # Define a refined color palette and hollow markers for clear differentiation
    colors = ['#4e79a7', '#f28e2b', '#e15759', '#76b7b2', '#59a14f', '#edc948', '#b07aa1', '#ff9da7', '#9c755f', '#bab0ab']  # Extended color-blind-friendly palette
    markers = ['o', 's', 'D', '^', 'v', 'P', 'X', '<', '>', '*']  # Unique markers for each dataset
    marker_color_combinations = itertools.cycle(zip(colors, markers))

    line_handles = []  # For legend without error bars
    for _set in data_sets_names:
        homophily_means = np.load(_set+'_homophily_means.npy', allow_pickle=True).item()
        homophily_lowers = np.load(_set+'_homophily_lowers.npy', allow_pickle=True).item()
        homophily_uppers = np.load(_set+'_homophily_uppers.npy', allow_pickle=True).item()

        # Use keys of the dictionaries as clique sizes
        #clique_sizes = list(homophily_means.keys())

        # Filter clique sizes <= 12
        clique_sizes = [c for c in homophily_means.keys() if c <= 12]
    
        homophily_values_list = [homophily_means[c] for c in clique_sizes]
        error_lower = [homophily_means[c] - homophily_lowers[c] for c in clique_sizes]
        error_upper = [homophily_uppers[c] - homophily_means[c] for c in clique_sizes]

        # Ensure error values are non-negative for plotting
        error_lower = [0 if el < 0 else el for el in error_lower]
        error_upper = [0 if eu < 0 else eu for eu in error_upper]

        # Get the next color and marker combination
        color, marker = next(marker_color_combinations)

        # Plot with error bars and hollow markers
        handle = plt.errorbar(
            clique_sizes,
            homophily_values_list,
            yerr=[error_lower, error_upper],
            color=color,
            linestyle='-',  # Use solid lines for all datasets
            linewidth=2,
            ecolor='gray',
            capsize=5,
            elinewidth=1.5,
            marker=marker,
            markerfacecolor='white',  # Hollow marker
            markeredgecolor=color,
            markeredgewidth=1.5,
            markersize=6
        )

        # Create a simplified legend handle without error bars
        label = label_map[_set] if label_map and _set in label_map else _set
        line_handles.append(plt.Line2D([0], [0], color=color, linestyle='-', marker=marker, markersize=6, 
                                       markerfacecolor='white', markeredgecolor=color, label=label))

    plt.xlabel('Clique Size', fontsize=14)
    plt.ylabel('Homophily $h_c$', fontsize=14)
    plt.xticks(ticks=range(min(clique_sizes), max(clique_sizes) + 2, 2), fontsize=12)  # Ensure integer ticks only
    plt.yticks(fontsize=12)
    plt.ylim(-0.7, 1.02)  # Extend y-axis for better separation
    #plt.grid(True, linestyle='--', alpha=0.3)  # Make grid very faded

    # Add legend without error bars
    plt.legend(handles=line_handles, fontsize=12, title_fontsize=14)

    plt.tight_layout()
    plt.savefig('all_boot.pdf')
    plt.show()

data_sets_names = ['conf_16', 'last_fm', 'mobile', 'Oberlin44', 'Pokec', 'cph']
label_map = {
    'conf_16': 'Conf 2016',
    'last_fm': 'Last.fm',
    'mobile': 'Mobile Data',
    'Oberlin44': 'Facebook',
    'Pokec': 'Pokec',
    'cph': 'CPH'
}
plot_(data_sets_names, label_map)
```

```python
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import itertools

# Set Seaborn style
sns.set_theme(style="whitegrid", font_scale=1.2)

def plot_(data_sets_names, label_map=None):
    plt.figure(figsize=(7, 5))  # Adjust figure size for better aspect ratio

    # Use Seaborn's color palette
    palette = sns.color_palette("tab10", n_colors=len(data_sets_names))
    markers = ['o', 's', 'D', '^', 'v', 'P', 'X', '<', '>', '*']
    marker_color_combinations = itertools.cycle(zip(palette, markers))

    line_handles = []  # For legend
    for _set in data_sets_names:
        homophily_means = np.load(_set + '_homophily_means.npy', allow_pickle=True).item()
        homophily_lowers = np.load(_set + '_homophily_lowers.npy', allow_pickle=True).item()
        homophily_uppers = np.load(_set + '_homophily_uppers.npy', allow_pickle=True).item()

        # Filter clique sizes  12
        clique_sizes = [c for c in homophily_means.keys() if c <= 12]
        homophily_values_list = [homophily_means[c] for c in clique_sizes]
        error_lower = [homophily_means[c] - homophily_lowers[c] for c in clique_sizes]
        error_upper = [homophily_uppers[c] - homophily_means[c] for c in clique_sizes]

        # Ensure error values are non-negative for plotting
        error_lower = [max(0, el) for el in error_lower]
        error_upper = [max(0, eu) for eu in error_upper]

        # Get the next color and marker combination
        color, marker = next(marker_color_combinations)

        # Plot main line
        plt.plot(
            clique_sizes, homophily_values_list,
            color=color, linestyle='-', linewidth=2.5,
            marker=marker, markersize=7,
            markerfacecolor='white', markeredgecolor=color, markeredgewidth=2,
            label=label_map.get(_set, _set)  # Use label_map if available
        )

        # Add transparent error band instead of error bars
        plt.fill_between(
            clique_sizes,
            np.array(homophily_values_list) - np.array(error_lower),
            np.array(homophily_values_list) + np.array(error_upper),
            color=color, alpha=0.2  # Transparency for the confidence interval
        )

    # Labels and axis formatting
    plt.xlabel('Clique Size', fontsize=14)
    plt.ylabel('Homophily $h_c$', fontsize=14)
    plt.xticks(ticks=range(2, 14, 2), fontsize=12)  # Set x-axis ticks for better readability
    plt.yticks(fontsize=12)
    plt.ylim(-0.7, 1.02)  # Keep y-axis limits for consistency
    plt.xlim(1, 12.5)  # Restrict x-axis to clique sizes 112

    # Legend outside plot for better clarity
    plt.legend(fontsize=12, title="Datasets", title_fontsize=14, loc="upper left", bbox_to_anchor=(1, 1))

    plt.tight_layout()  # Prevent label cutoff
    plt.savefig('all_boot.pdf', dpi=300, bbox_inches='tight')  # High-res save
    plt.show()

# Dataset names and labels
data_sets_names = ['conf_16', 'last_fm', 'mobile', 'Oberlin44', 'Pokec', 'cph']
label_map = {
    'conf_16': 'Conf 2016',
    'last_fm': 'Last.fm',
    'mobile': 'Mobile Data',
    'Oberlin44': 'Oberlin FB',
    'Pokec': 'Pokec',
    'cph': 'CPH'
}

# Call function
plot_(data_sets_names, label_map)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path

def plot_pokec_transition_nice():
    # Directory containing bootstrap .npy files
    boots_dir = Path("/home/karimia1/Weak-and-Strong-Homophily-in-Networks/real_world_data/boots")
    
    # File paths
    mean_file  = boots_dir / "Pokec_homophily_means.npy"
    lower_file = boots_dir / "Pokec_homophily_lowers.npy"
    upper_file = boots_dir / "Pokec_homophily_uppers.npy"
    
    # Load data
    means  = np.load(mean_file, allow_pickle=True).item()
    lowers = np.load(lower_file, allow_pickle=True).item()
    uppers = np.load(upper_file, allow_pickle=True).item()
    
    # Prepare data for clique sizes up to 10
    sizes   = sorted([c for c in means if c <= 10])
    h_vals  = np.array([means[c]           for c in sizes])
    err_low = np.array([means[c] - lowers[c] for c in sizes])
    err_up  = np.array([uppers[c] - means[c] for c in sizes])
    yerr    = np.vstack([err_low, err_up])
    
    # Plot setup
    fig, ax = plt.subplots(figsize=(6, 4))
    ax.set_ylim(-0.7, 0.45)
    ax.set_xlim(1.5, 10.5)
    
    # Shade heterophily vs. homophily regions
    ax.axhspan(-0.7, 0, color='salmon', alpha=0.2)
    ax.axhspan(0, 0.45, color='lightseagreen', alpha=0.2)
    
    # Plot error bars
    ax.errorbar(
        sizes, h_vals, yerr=yerr,
        fmt='o-', capsize=4, linewidth=2.5, markersize=7,
        markerfacecolor='white', markeredgewidth=2,
        markeredgecolor='darkblue', color='darkblue'
    )
    
    # Dashed reference line at 0
    ax.axhline(0, color='black', linestyle='--', linewidth=1)
    
    # Axis labels and style
    ax.set_xlabel("Clique size", fontsize=14)
    ax.set_ylabel("Homophily $h_c$", fontsize=14)
    ax.set_xticks(sizes)
    ax.tick_params(axis='both', labelsize=12)
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    
    plt.tight_layout()
    
    # Save figure back into boots directory
    out_path = boots_dir / "pokec_homophily_transition_nice.pdf"
    plt.savefig(out_path, dpi=300, bbox_inches="tight")
    plt.show()
    print(f"Figure saved to {out_path}")

# Execute the plotting function
plot_pokec_transition_nice()
```

```python
import numpy as np

def generate_latex_table(data_sets_names, label_map=None):
    latex_output = ""

    for _set in data_sets_names:
        homophily_means = np.load(_set + '_homophily_means.npy', allow_pickle=True).item()
        homophily_lowers = np.load(_set + '_homophily_lowers.npy', allow_pickle=True).item()
        homophily_uppers = np.load(_set + '_homophily_uppers.npy', allow_pickle=True).item()

        dataset_name = label_map.get(_set, _set)  # Use label_map if available

        latex_output += f"\\begin{{table}}[h]\n\\centering\n"
        latex_output += f"\\caption{{Homophily values $h_c$ for {dataset_name}}}\n"
        latex_output += "\\begin{tabular}{c|c}\n"
        latex_output += "\\hline\n"
        latex_output += "Clique Size $c$ & Homophily $h_c$ \\\\ \n"
        latex_output += "\\hline\n"

        for c in sorted(homophily_means.keys()):
            hc = homophily_means[c]
            error_lower = hc - homophily_lowers[c]
            error_upper = homophily_uppers[c] - hc
            latex_output += f"{c} & ${hc:.2f}_{{-{error_lower:.2f}}}^{{+{error_upper:.2f}}}$ \\\\ \n"

        latex_output += "\\hline\n"
        latex_output += "\\end{tabular}\n"
        latex_output += f"\\label{{tab:{_set}_homophily}}\n"
        latex_output += "\\end{table}\n\n"

    return latex_output

# Dataset names and labels
data_sets_names = ['conf_16', 'Pokec', 'Oklahoma97', 'mobile', 'last_fm', 'cph']

label_map = {
    'conf_16': 'Conf 2016',
    'last_fm': 'Last.fm',
    'mobile': 'Mobile Data',
    'Oberlin44': 'Oberlin FB',
    'Pokec': 'Pokec',
    'cph': 'CPH',
    'Oklahoma97': 'Oklahoma FB'
}

# Generate and print LaTeX tables
latex_tables = generate_latex_table(data_sets_names, label_map)
print(latex_tables)
```
